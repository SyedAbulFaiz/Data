**Matplotlib**

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(0,100,7)
x
```

```python
y=x*2
y
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.plot(x,y,color='pink',linestyle=':',linewidth=7)
plt.grid(True)
```

```python
x
```

```python
y1=2*x
y2=3*x
y1
```

```python
y2
```

*Adding two lines in the same plot*

```python
plt.plot(x,y1,color='pink',linewidth=2)
plt.plot(x,y2,color='purple',linewidth=2)
plt.grid(True)
```

*Adding sub-plots*

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='Blue',linestyle=':',linewidth=2)
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(x,y2,color='Red',linestyle=':',linewidth=2)
plt.grid(True)
```

**bar plot**

```python
student = {'Bonn':45,'Kiel':32,'Stuttgart':75}
names = list(student.keys()) 
Marks = list(student.values())
plt.bar(names,Marks,color='pink')
plt.grid(True)
```

 *Adding titles and labels*

```python
plt.bar(names,Marks,color='pink')
plt.title('Student Result')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
```

**horizontal bar plot**

```python
plt.barh(names,Marks,color='blue')
plt.title('Student Result')
plt.xlabel('Marks')
plt.ylabel('Names')
plt.grid(True)
```

*Scatter plot*

```python
x=np.random.randint(1,25,22)
y=np.random.randint(2,24,22)
plt.scatter(x,y)
plt.grid(True)
```

```python
#changing mark aesthetics
x=np.random.randint(1,25,22)
y=np.random.randint(2,24,22)
plt.scatter(x,y,marker='*',c='r',s=100)
```

```python
z=np.random.randint(1,30,22)
```

*Adding two markers in the same plot*

```python
plt.scatter(x,y,marker='*',c='r',s=50)
plt.scatter(x,z,marker='.',c='b',s=150)
```

*Adding subplots*

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',c='r',s=50)
plt.subplot(1,2,2)
plt.scatter(x,z,marker='.',c='b',s=150)
```

*Pie Chart*

**SEABORN**

```python
import seaborn as sns
from matplotlib import pyplot as plt 
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
```
